### All sales

```python
sales_containers = driver.find_elements(By.CSS_SELECTOR, "div._6uN7R")
```

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏•‡∏¥‡∏™‡∏ï‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
sales_data = []

# ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏†‡∏≤‡∏¢‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞ container
for container in sales_containers:
    try:
        # ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤ span ‡∏ó‡∏µ‡πà‡∏°‡∏µ class '_1cEkb'
        sales_span = container.find_element(By.CSS_SELECTOR, "span._1cEkb > span")
        sales_text = sales_span.text.strip()  # ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏•‡∏∞‡∏•‡∏ö‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
        if sales_text:
            sales_data.append(sales_text)  # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏•‡∏á‡πÉ‡∏ô‡∏•‡∏¥‡∏™‡∏ï‡πå
        else:
            sales_data.append("none")  # ‡∏ñ‡πâ‡∏≤‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ß‡πà‡∏≤‡∏á ‡πÉ‡∏´‡πâ‡πÉ‡∏™‡πà "none"
    except:
        sales_data.append("none")  # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏û‡∏ö <span> ‡πÉ‡∏™‡πà "none"

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
for index, sales in enumerate(sales_data, start=1):
    print(f"‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà {index}: {sales}")
```

### All review

```python
review_containers = driver.find_elements(By.CSS_SELECTOR, "div._6uN7R")
```

```python
reviews_data = []

# ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞ container
for container in review_containers:
    try:
        # ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤ span ‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏£‡∏µ‡∏ß‡∏¥‡∏ß
        review_span = container.find_element(By.CSS_SELECTOR, "span.qzqFw")
        review_text = review_span.text.strip()  # ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏•‡∏∞‡∏•‡∏ö‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á
        if review_text:
            reviews_data.append(review_text)  # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏•‡∏á‡πÉ‡∏ô‡∏•‡∏¥‡∏™‡∏ï‡πå
        else:
            reviews_data.append("none")  # ‡∏ñ‡πâ‡∏≤‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ß‡πà‡∏≤‡∏á ‡πÉ‡∏™‡πà "none"
    except:
        reviews_data.append("none")  # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏û‡∏ö <span> ‡πÉ‡∏™‡πà "none"

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
for index, review in enumerate(reviews_data, start=1):
    print(f"‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà {index}: {review}")
```

```python
reviews_series = pd.Series(reviews_data)
reviews_series = reviews_series.astype(str).apply(lambda x: re.sub(r'[()]', '', x))
print(reviews_series.head())
```

### All rating

```python
rating_containers = driver.find_elements(By.CSS_SELECTOR, "div._6uN7R")
```

```python
ratings = []
for container in rating_containers:
    stars = container.find_elements(By.CSS_SELECTOR, "i._9-ogB.Dy1nx")  # ‡πÉ‡∏ä‡πâ class ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏î‡∏≤‡∏ß
    if len(stars) > 0:
        ratings.append(len(stars))  # ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏î‡∏≤‡∏ß‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞ container
    else:
        ratings.append("none")  # ‡πÉ‡∏™‡πà "none" ‡∏´‡∏≤‡∏Å‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏î‡∏≤‡∏ß

# ‡πÅ‡∏™‡∏î‡∏á‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏î‡∏≤‡∏ß‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£
for index, rating in enumerate(ratings, start=1):
    print(f"‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà {index}: {rating} ‡∏î‡∏≤‡∏ß")
```

### All locations

```python
all_locations = soup.find_all('span',{'class':'oa6ri'})
```

```python
all_locations_list = []
for locations in all_locations:
    all_locations_list.append(locations.text)
    
df_locations = pd.DataFrame(all_locations_list, columns=['locations'])
print(df_locations.head())
```

### All images

```python
images = driver.find_elements(By.CSS_SELECTOR, "div.picture-wrapper.jBwCF img")
```

```python
image_urls = []
for img in images:
    image_urls.append(img.get_attribute("src"))

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• URL ‡∏Ç‡∏≠‡∏á‡∏£‡∏π‡∏õ‡∏†‡∏≤‡∏û
print("Image URLs:")
for url in image_urls:
    print(url)
```

### All Data

```python
all_data_list = pd.DataFrame([all_product_list,all_price_list,sales_data,ratings,reviews_series,df_locations['locations'],image_urls])
```

```python
all_data_list = all_data_list.transpose()
all_data_list.columns = ['title','price','amount','amount_ratings','amount_review','locations','image_urls']
```
