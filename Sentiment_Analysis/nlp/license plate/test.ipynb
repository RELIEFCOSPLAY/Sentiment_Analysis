```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, img_as_float
from numpy.fft import fft2, ifft2, fftshift

# Load the image and convert to grayscale
img0 = io.imread('License_plate_003.jpg')
img1 = color.rgb2gray(img0)
img = img1[1:256, 30:285]  # Equivalent to MATLAB's imcrop(img1,[30 1 255 255])
img = img_as_float(img)

# Fourier transform of the image
spect = fft2(img, s=(256, 256))
shiftspect = fftshift(spect)

# Plot original image
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

#===================================================================
# Low pass filtering (LPF) with different mask sizes
#===================================================================
# Medium LPF mask
LPmaskM = np.zeros((256, 256))
LPmaskM[113:144, 113:144] = 1

# Apply LPF mask
fil = shiftspect * LPmaskM
fil2 = fftshift(fil)
output2 = ifft2(fil2)
output2 = np.abs(output2)

# Display filtered image
plt.subplot(3, 2, 2)
plt.imshow(output2, cmap='gray')
plt.title('Medium Low Pass Filtered Image')

# Small LPF mask
LPmaskS = np.zeros((256, 256))
LPmaskS[97:160, 97:160] = 1

# Apply LPF mask
fil = shiftspect * LPmaskS
fil2 = fftshift(fil)
output1 = ifft2(fil2)
output1 = np.abs(output1)

# Display filtered image
plt.subplot(3, 2, 4)
plt.imshow(output1, cmap='gray')
plt.title('Small Low Pass Filtered Image')

# Large LPF mask
LPmaskL = np.zeros((256, 256))
LPmaskL[87:170, 87:170] = 1

# Apply LPF mask
fil = shiftspect * LPmaskL
fil2 = fftshift(fil)
output3 = ifft2(fil2)
output3 = np.abs(output3)

# Display filtered image
plt.subplot(3, 2, 6)
plt.imshow(output3, cmap='gray')
plt.title('Large Low Pass Filtered Image')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float
from numpy.fft import fft2, ifft2, fftshift, ifftshift

# Load the image (already in grayscale)
img = io.imread('License_plate_16_gray_levels.jpg')  # Replace with your image file path
img = img_as_float(img)  # Use the entire image without cropping

# Get the size of the image
rows, cols = img.shape

# Fourier transform of the image
spect = fft2(img)
shiftspect = fftshift(spect)

# Plot original image
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

#===================================================================
# Low pass filtering (LPF) with different mask sizes
#===================================================================
# Medium LPF mask
LPmaskM = np.zeros((rows, cols))
LPmaskM[rows//2 - 15:rows//2 + 15, cols//2 - 15:cols//2 + 15] = 1

# Apply LPF mask
fil = shiftspect * LPmaskM
fil2 = ifftshift(fil)
output2 = ifft2(fil2)
output2 = np.abs(output2)

# Display filtered image
plt.subplot(3, 2, 2)
plt.imshow(output2, cmap='gray')
plt.title('Medium Low Pass Filtered Image')

# Small LPF mask
LPmaskS = np.zeros((rows, cols))
LPmaskS[rows//2 - 30:rows//2 + 30, cols//2 - 30:cols//2 + 30] = 1

# Apply LPF mask
fil = shiftspect * LPmaskS
fil2 = ifftshift(fil)
output1 = ifft2(fil2)
output1 = np.abs(output1)

# Display filtered image
plt.subplot(3, 2, 4)
plt.imshow(output1, cmap='gray')
plt.title('Small Low Pass Filtered Image')

# Large LPF mask
LPmaskL = np.zeros((rows, cols))
LPmaskL[rows//2 - 40:rows//2 + 40, cols//2 - 40:cols//2 + 40] = 1

# Apply LPF mask
fil = shiftspect * LPmaskL
fil2 = ifftshift(fil)
output3 = ifft2(fil2)
output3 = np.abs(output3)

# Display filtered image
plt.subplot(3, 2, 6)
plt.imshow(output3, cmap='gray')
plt.title('Large Low Pass Filtered Image')

plt.tight_layout()
plt.show()

#===================================================================
# High pass filtering (HPF) with different mask sizes
#===================================================================
plt.figure(figsize=(12, 8))

# Medium HPF mask
HPmaskM = np.ones((rows, cols))
HPmaskM[rows//2 - 15:rows//2 + 15, cols//2 - 15:cols//2 + 15] = 0

# Apply HPF mask
filh = shiftspect * HPmaskM
filh2 = ifftshift(filh)
output2 = ifft2(filh2)
output2 = np.abs(output2)

# Display filtered image
plt.subplot(3, 2, 2)
plt.imshow(output2, cmap='gray')
plt.title('Medium High Pass Filtered Image')

# Small HPF mask
HPmaskS = np.ones((rows, cols))
HPmaskS[rows//2 - 30:rows//2 + 30, cols//2 - 30:cols//2 + 30] = 0

# Apply HPF mask
filh = shiftspect * HPmaskS
filh2 = ifftshift(filh)
output1 = ifft2(filh2)
output1 = np.abs(output1)

# Display filtered image
plt.subplot(3, 2, 4)
plt.imshow(output1, cmap='gray')
plt.title('Small High Pass Filtered Image')

# Large HPF mask
HPmaskL = np.ones((rows, cols))
HPmaskL[rows//2 - 40:rows//2 + 40, cols//2 - 40:cols//2 + 40] = 0

# Apply HPF mask
filh = shiftspect * HPmaskL
filh2 = ifftshift(filh)
output3 = ifft2(filh2)
output3 = np.abs(output3)

# Display filtered image
plt.subplot(3, 2, 6)
plt.imshow(output3, cmap='gray')
plt.title('Large High Pass Filtered Image')

plt.tight_layout()
plt.show()

#===================================================================
# Band pass filtering (BPF)
#===================================================================
plt.figure(figsize=(12, 8))

# Medium BPF mask
BPmaskM = np.zeros((rows, cols))
BPmaskM[rows//2 - 85:rows//2 + 85, cols//2 - 85:cols//2 + 85] = 1
BPmaskM[rows//2 - 15:rows//2 + 15, cols//2 - 15:cols//2 + 15] = 0

# Apply BPF mask
fil = shiftspect * BPmaskM
fil2 = ifftshift(fil)
output2 = ifft2(fil2)
output2 = np.abs(output2)

# Display filtered image
plt.subplot(2, 2, 2)
plt.imshow(output2, cmap='gray')
plt.title('Medium Band Pass Filtered Image')

# Small BPF mask
BPmaskS = np.zeros((rows, cols))
BPmaskS[rows//2 - 95:rows//2 + 95, cols//2 - 95:cols//2 + 95] = 1
BPmaskS[rows//2 - 30:rows//2 + 30, cols//2 - 30:cols//2 + 30] = 0

# Apply BPF mask
fil = shiftspect * BPmaskS
fil2 = ifftshift(fil)
output1 = ifft2(fil2)
output1 = np.abs(output1)

# Display filtered image
plt.subplot(2, 2, 4)
plt.imshow(output1, cmap='gray')
plt.title('Small Band Pass Filtered Image')

plt.tight_layout()
plt.show()
```
